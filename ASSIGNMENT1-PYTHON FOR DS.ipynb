Visualization

```python
plt.figure(figsize=(8,8))
plt.hist(iris_data['SL'],color='violet',rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('Sepal Length',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(iris_data['SW'],color='black',rwidth=0.9)
plt.title('Distribution of SW',fontsize=16)
plt.xlabel('Sepal Width',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(iris_data['PL'],color='red',rwidth=0.9)
plt.title('Distribution of PL',fontsize=16)
plt.xlabel('Petal Length',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(iris_data['PW'],color='green',rwidth=0.9)
plt.title('Distribution of PW',fontsize=16)
plt.xlabel('Petal Width',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(iris_data['SL'],iris_data['SW'],s=10,marker='^',c='red')
plt.title('Scatter plot of Sepal Lenth Vs Sepal Width',fontsize=16)
plt.xlabel('Sepal Length',fontsize=14)
plt.ylabel('Sepal Width',fontsize=14)


plt.subplot(2,1,2)
plt.scatter(iris_data['PL'],iris_data['PW'],s=10,marker='^',c='green')
plt.title('Scatter plot of Petal Length Vs Petal Width',fontsize=16)
plt.xlabel('Petal Length',fontsize=14)
plt.ylabel('Petal Width',fontsize=14)
plt.tight_layout()
```

Pair Plot

```python
sns.pairplot(iris_data)
```

Pair plot is a type of plot available in Seaborn Library and it is very useful to undestand the relationship between all the 
variables present in the data.It gives Scatter plot aswell as frequency distribution in a singleframe.Scatter plot is mainly 
used to know whether there is any relation between varibles or not.
